Setup and Upload Your Data

```python
# Cell 1: Setup and Load Your Data (Improved)

import pandas as pd
import matplotlib.pyplot as plt
import os
from openpyxl import load_workbook

# --- IMPORTANT: Set your Excel file path here ---
# Option 1: If LiveData.xlsx is in the same directory as this Jupyter Notebook
excel_file_path = 'LiveData.xlsx' 

# Option 2: If LiveData.xlsx is in a specific folder (e.g., F:\GP\ML\LiveData\)
# Make sure to use double backslashes or forward slashes for paths in Python strings
excel_file_path = r'F:\GP\ML\LiveData\LiveData.xlsx' 

# List to hold individual DataFrames for each session, with cumulative time
sessions_data = [] 
cumulative_time_offset = 0

if not os.path.exists(excel_file_path):
    print(f"Error: Excel file not found at '{excel_file_path}'. Please check the path.")
else:
    print(f"Attempting to load data from: '{excel_file_path}'")
    
    try:
        temp_wb = load_workbook(excel_file_path)
        sheet_names_to_read = [s for s in temp_wb.sheetnames if s.startswith("Session_")]
        
        if not sheet_names_to_read:
            print(f"No 'Session_' sheets found in '{excel_file_path}'. Please ensure your data sheets start with 'Session_'.")
        else:
            print(f"Found sheets starting with 'Session_': {sheet_names_to_read}")
            for sheet_name in sheet_names_to_read:
                try:
                    sheet_df = pd.read_excel(excel_file_path, sheet_name=sheet_name)
                    
                    # --- OPTIONAL: Filter out initial spike / bad data ---
                    # Remove the very first row (index 0) if it often causes large, artifactual spikes.
                    # Only apply if the sheet has more than one row.
                    if len(sheet_df) > 1:
                        sheet_df = sheet_df.iloc[1:].copy() 
                    else:
                        print(f"Skipping sheet '{sheet_name}' as it has 1 or fewer rows after potential header/spike removal.")
                        continue # Skip to the next sheet if it becomes empty
                    
                    # Ensure 'Time (s)' column exists for time-series plotting
                    if 'Time (s)' not in sheet_df.columns:
                        print(f"Skipping sheet '{sheet_name}': 'Time (s)' column not found.")
                        continue

                    # Calculate cumulative time for this session
                    sheet_df['Cumulative_Time (s)'] = sheet_df['Time (s)'] + cumulative_time_offset
                    sessions_data.append({'name': sheet_name, 'df': sheet_df})
                    
                    # Update offset for the next session
                    if not sheet_df.empty:
                        cumulative_time_offset = sheet_df['Cumulative_Time (s)'].iloc[-1]
                    # If sheet_df is empty here, cumulative_time_offset remains the same for next session

                    print(f"Successfully loaded sheet '{sheet_name}'. Rows for plotting: {len(sheet_df)}")
                except Exception as e:
                    print(f"Error reading sheet '{sheet_name}': {e}")
                
        if not sessions_data: # If no valid sheets were loaded at all
            print("No valid data loaded from any 'Session_' sheet after processing. Please check your Excel file and sheet names.")
        else:
            print(f"\nAll valid 'Session_' sheets loaded for plotting.")
            print(f"Total number of sessions loaded: {len(sessions_data)}")

            # Display columns from the first loaded session as a reference
            if sessions_data:
                print("\nAvailable columns for plotting (from the first loaded session, copy these names for Cell 3):")
                print(sessions_data[0]['df'].columns.tolist())
            
    except Exception as e:
        print(f"Error loading Excel file '{excel_file_path}': {e}")
        print("Please ensure the file is a valid .xlsx format and not corrupted/locked.")

# Define df_combined as an empty DataFrame or the actual combined one if needed elsewhere
# For plotting, we'll iterate through sessions_data
df_combined_for_info = pd.concat([s['df'] for s in sessions_data]) if sessions_data else pd.DataFrame()
if not df_combined_for_info.empty:
    print(f"\nTotal combined rows (for info, not direct plotting): {len(df_combined_for_info)}")
```

Define the Plotting Function


```python
# Cell 2: Define the Plotting Function

def plot_features_by_session(sessions_list, features_to_plot, title="Eye Movement Features", x_column='Cumulative_Time (s)', figsize=(12, 6)):
    """
    Plots selected features from multiple sessions as line charts, with each session
    plotted as a distinct segment on a potentially cumulative time axis.

    Args:
        sessions_list (list): A list of dictionaries, each containing {'name': str, 'df': pd.DataFrame}.
                              Each DataFrame corresponds to a single session.
        features_to_plot (list): A list of column names (strings) to plot.
        title (str): The title of the plot.
        x_column (str): The column name to use for the x-axis (e.g., 'Cumulative_Time (s)').
        figsize (tuple): Figure size for the plot (width, height).
    """
    
    if not sessions_list:
        print("No sessions data provided for plotting.")
        return

    plt.figure(figsize=figsize)
    plotted_any = False

    for session_info in sessions_list:
        session_name = session_info['name']
        session_df = session_info['df']

        if session_df.empty:
            continue
            
        if x_column not in session_df.columns:
            print(f"Error: X-axis column '{x_column}' not found in DataFrame for session '{session_name}'. Skipping session.")
            continue

        for feature in features_to_plot:
            if feature in session_df.columns:
                plt.plot(session_df[x_column], session_df[feature], label=f'{session_name} - {feature}')
                plotted_any = True
            else:
                print(f"Warning: Feature '{feature}' not found in DataFrame for session '{session_name}'. Skipping.")

    if plotted_any:
        plt.title(title)
        plt.xlabel(x_column)
        plt.ylabel("Value")
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside for clarity
        plt.grid(True)
        plt.tight_layout() # Adjusts plot parameters for a tight layout
        plt.show()
    else:
        print("No valid features were found to plot across any session.")
```
